```python


# Lire le fichier en utilisant le délimiteur ';' et en précisant que la première ligne est l'en-tête
cinema = pd.read_csv('/Users/sara/Desktop/Albert School/HTML/Cinema/les_salles_de_cinemas_en_ile-de-france 3.csv', delimiter=';', header=1)

# Afficher les premières lignes pour vérifier
cinema.head()
```

```python
import matplotlib.pyplot as plt

# Comptage des cinémas par département
departement_counts = cinema['Département'].value_counts()

# Diagramme en barres
plt.figure(figsize=(10, 6))
departement_counts.plot(kind='bar')
plt.title("Nombre de salles de cinéma par département")
plt.xlabel("Département")
plt.ylabel("Nombre de salles")
plt.show()
```

```python
# Nuage de points pour les écrans et fauteuils
plt.figure(figsize=(10, 6))
plt.scatter(cinema['écrans'], cinema['fauteuils'], alpha=0.5)
plt.title("Relation entre le nombre d'écrans et de fauteuils par cinéma")
plt.xlabel("Nombre d'écrans")
plt.ylabel("Nombre de fauteuils")
plt.show()
```

```python
# Comptage des tranches d'entrées
tranche_counts = cinema['Tranche d\'entrées annuelles'].value_counts()

# Diagramme en barres
plt.figure(figsize=(10, 6))
tranche_counts.plot(kind='bar')
plt.title("Répartition des tranches d'entrées annuelles")
plt.xlabel("Tranche d'entrées")
plt.ylabel("Nombre de cinémas")
plt.show()
```

```python
# Sélection des colonnes d'entrées pour 2019 et 2020
entrées = cinema[['entrées 2020', 'entrées 2019']].dropna()

# Diagramme en barres
plt.figure(figsize=(10, 6))
plt.hist([entrées['entrées 2019'], entrées['entrées 2020']], label=['Entrées 2019', 'Entrées 2020'], bins=20)
plt.title("Evolution des entrées entre 2019 et 2020")
plt.xlabel("Nombre d'entrées")
plt.ylabel("Nombre de cinémas")
plt.legend()
plt.show()
```
