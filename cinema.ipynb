```python

import os
import pandas as pd 


pd.options.display.max_columns = 200
pd.options.display.max_rows = 3000
```

```python
cinema_data = pd.read_csv('les_salles_de_cinemas_en_ile-de-france 3.csv', delimiter=';', header=1)
```

```python
cinema_data.head()
```

```python
print(cinema_data.columns)
```

```python
# 1. Supprimer les doublons
cinema_data.drop_duplicates(inplace=True)

# 2. Remplacer les valeurs manquantes pour certaines colonnes avec des valeurs par défaut
cinema_data['programmateur'].fillna('Inconnu', inplace=True)
cinema_data['label Art et Essai'].fillna('Inconnu', inplace=True)

# 3. Remplacer les valeurs manquantes dans les colonnes numériques avec la moyenne ou la médiane
# Par exemple, pour les entrées annuelles:
cinema_data['entrées 2019'].fillna(cinema_data['entrées 2019'].mean(), inplace=True)
cinema_data['entrées 2020'].fillna(cinema_data['entrées 2020'].mean(), inplace=True)

# 4. Vérifier les valeurs aberrantes dans certaines colonnes, par exemple les écrans ou les fauteuils
# Suppression des valeurs hors limites (facultatif, selon le besoin)
cinema_data = cinema_data[(cinema_data['écrans'] > 0) & (cinema_data['fauteuils'] > 0)]

# 5. Conversion des colonnes de pourcentage (ex. "PDM en entrées des films français 2020") en décimales si besoin
percentage_columns = [
    'PdM en entrées des films français 2020',
    'PdM en entrées des films américains 2020',
    'PdM en entrées des films européens 2020',
    'PdM en entrées des autres films 2020',
    'PdM en entrées des films Art et Essai 2020'
]
for col in percentage_columns:
    cinema_data[col] = cinema_data[col] / 100  # Diviser par 100 pour avoir des valeurs entre 0 et 1

# 6. Séparer les données de géolocalisation en latitude et longitude
cinema_data[['latitude', 'longitude']] = cinema_data['geo'].str.split(', ', expand=True).astype(float)

# Afficher le dataset nettoyé
cinema_data.head()
```
