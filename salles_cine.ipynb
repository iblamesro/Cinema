Introduction : 

Le cinéma est un art et une industrie qui joue un rôle majeur dans la culture et le divertissement mondial. En France, il occupe une place centrale, notamment en Île-de-France, où se trouvent une grande majorité des salles de projection du pays. Cette région, grâce à sa densité de population et son rôle économique, constitue un pôle essentiel pour l’industrie cinématographique.


À travers cette étude, nous analysons le paysage cinématographique en Île-de-France, en examinant des aspects tels que la répartition géographique des salles, leurs caractéristiques (écrans, fauteuils) et leur fréquentation annuelle. Cela nous permet d’appréhender la diversité et l'impact du cinéma dans cette région.

```python
import os
import pandas as pd 


pd.options.display.max_columns = 200
pd.options.display.max_rows = 3000
```

```python
# Calcul de la moyenne annuelle des entrées par cinéma
moyenne_annuelle_par_cinema = cinema_data.groupby('nom')['entrées_annuelles'].mean()

# Affichage de la moyenne annuelle des entrées par cinéma
print("Moyenne annuelle des entrées par cinéma :\n")
print(moyenne_annuelle_par_cinema)

```

```python
# Calcul de la moyenne des fauteuils
moyenne_fauteuils = cinema_data['fauteuils'].mean()

print(f'Moyenne du nombre de fauteuils par cinéma : {moyenne_fauteuils}')
```

```python
# Compter le nombre de cinémas par département
cinémas_par_département = cinema_data['Département'].value_counts()

print('Nombre de cinémas par département:')
print(cinémas_par_département)

# Visualisation
cinémas_par_département.plot(kind='bar', figsize=(10, 6))
plt.title('Nombre de cinémas par département')
plt.xlabel('Département')
plt.ylabel('Nombre de cinémas')
plt.show()
```

```python
import seaborn as sns

# Calculer la corrélation entre le nombre d'écrans et la fréquentation
sns.scatterplot(data=cinema_data, x='écrans', y='entrées_annuelles')
plt.title('Corrélation entre le nombre d\'écrans et la fréquentation')
plt.xlabel('Nombre d\'écrans')
plt.ylabel('Entrées annuelles')
plt.show()

# Afficher la corrélation numérique
corr = cinema_data[['écrans', 'entrées_annuelles']].corr()
print(f'Corrélation entre les écrans et les entrées annuelles : \n{corr}')
```

```python
# Calculer la corrélation entre le nombre de fauteuils et la fréquentation
sns.scatterplot(data=cinema_data, x='fauteuils', y='entrées_annuelles')
plt.title('Corrélation entre le nombre de fauteuils et la fréquentation')
plt.xlabel('Nombre de fauteuils')
plt.ylabel('Entrées annuelles')
plt.show()

# Afficher la corrélation numérique
corr_fauteuils = cinema_data[['fauteuils', 'entrées_annuelles']].corr()
print(f'Corrélation entre les fauteuils et les entrées annuelles : \n{corr_fauteuils}')
```

```python
print(cinema_data.columns)
```

```python


# Si la colonne "propriétaire" est présente, faire l'agrégation pour les comparaisons
if 'propriétaire' in cinema_data.columns:
    # Comparer les caractéristiques des cinémas indépendants vs chaînes
    indépendants_vs_chaines = cinema_data.groupby('propriétaire').agg({
        'fauteuils': 'mean',
        'écrans': 'mean',
        'entrées_annuelles': 'mean'
    })

    print('Comparaison des cinémas indépendants vs chaînes :')
    print(indépendants_vs_chaines)
else:
    print("La colonne 'propriétaire' est manquante. Vérifiez l'orthographe des colonnes.")
```

```python
moyenne_entrées = cinema_data['entrées_annuelles'].mean()
print(f"Moyenne des entrées annuelles par cinéma : {moyenne_entrées}")
```

```python
# Trouver les cinémas avec les entrées annuelles les plus élevées et les plus faibles
cinema_max = cinema_data.loc[cinema_data['entrées_annuelles'].idxmax()]
cinema_min = cinema_data.loc[cinema_data['entrées_annuelles'].idxmin()]

# Afficher le nom du cinéma et la ville pour chaque cas
print(f"Cinéma avec le plus grand nombre d'entrées annuelles : {cinema_max['nom']} à {cinema_max['commune']}")
print(f"Cinéma avec le plus petit nombre d'entrées annuelles : {cinema_min['nom']} à {cinema_min['commune']}")
```

```python
# Calcul de l'évolution des entrées entre 2019 et 2020
cinema_data['évolution entrées'] = (cinema_data['entrées 2020'] - cinema_data['entrées 2019']) / cinema_data['entrées 2019'] * 100
print(cinema_data[['nom', 'commune', 'évolution entrées']].sort_values(by='évolution entrées', ascending=False))
```
